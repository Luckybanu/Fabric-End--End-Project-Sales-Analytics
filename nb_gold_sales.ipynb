```sql
CREATE VIEW gold_fact_sales AS
SELECT
    f.*,                                    -- all columns from silver_fact_sales
    fx.rate                      AS fx_rate_usd,
    (f.gross_amount * fx.rate)   AS gross_amount_usd
FROM silver_fact_sales f
JOIN silver_fx_rates fx
      ON fx.currency = f.currency
     AND fx.date     = CAST(f.order_date AS date)   -- align types
WHERE f.order_date >= '2024-01-01';
```

```sql
SELECT TOP 10 *
FROM gold_fact_sales;
```
